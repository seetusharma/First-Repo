```python
from pandas import * 
from datetime import *
from numpy import *
```

```python
f=pd.read_csv(r"Desktop\new1.csv",names=['Sr. No.','POSITION','DOB','FLAG','GENDER','DOJ'],header=0)
f.head()
```

```python
today=to_datetime('today')
f['DOB']=to_datetime(f['DOB'])
AG=(today-f['DOB'])/timedelta64(1,'Y')
f['Age']=round(AG)
f['DOJ']=to_datetime(f['DOJ'])
Exp=(today-f['DOJ'])/timedelta64(1,'Y')
f['Experience']=round(Exp)
f['Seniority']=f[f['Experience']>=10]
f


```
